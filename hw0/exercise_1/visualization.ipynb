```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
comparisions_df = pd.read_csv('data/comparisons.csv', index_col=0)
swaps_df = pd.read_csv('data/swaps.csv', index_col=0)
time_df = pd.read_csv('data/time.csv', index_col=0)
```

```python
time_df.head()
```

```python
comparisions_df.head()
```

```python
swaps_df.head()
```

```python
# 1. t tn cho index  khi reset n thnh ct s c tn
time_df.index.name = 'Input Size'

# 2. Reset index  bin 'Input Size' thnh mt ct d liu
df_reset = time_df.reset_index()

# 3. Chuyn i d liu sang dng "Long Form" cho Seaborn
# id_vars l ct trc X, cc ct cn li s thnh 'Algorithm' v 'Time'
df_melted = df_reset.melt(id_vars=['Input Size'], var_name='Algorithm', value_name='Time')

# 4. V biu 
plt.figure(figsize=(12, 8))
# sns.set_theme(style="whitegrid")

sns.lineplot(
    data=df_melted, 
    x='Input Size', 
    y='Time', 
    hue='Algorithm',    # Mu khc nhau cho tng thut ton
    style='Algorithm',  # Kiu nt khc nhau (nt t/lin)
    markers=True,       # C du chm trn ti cc im
    dashes=False,       #  ng lin mch (ty chn)
    linewidth=2.5
)

plt.title('Thi gian thc thi cc thut ton sp xp', fontsize=16)
plt.ylabel('Thi gian thc thi (s)', fontsize=12)
plt.xlabel('Ln thc hin', fontsize=12)
plt.legend(title='Thut ton', fontsize=10)
plt.savefig('out/sorting_algorithms_time_comparison.png',bbox_inches='tight')
plt.show()
```
